```python
from transformers import pipeline
pipe = pipeline("image-classification", "umm-maybe/AI-image-detector")
```

```python
def image_classifier(image):
    outputs = pipe(image)
    results = {}
    for result in outputs:
        results[result['label']] = result['score']
    if results['human']>results['artificial']:
        return "Real"
    else:
        return "AI"
```

```python
from PIL import Image
import os

df = {}
for img in os.walk("/content/data/Data/Test_Images"):
    for img2 in img[2]:
      image = Image.open("/content/data/Data/Test_Images/"+ img2)
      # print(image_classifier(image))
      df[img2[:-4]] = image_classifier(image)

df = pd.DataFrame(list(df.items()), columns = ["Id", "Label"])
df.to_csv("predictions.csv", index=False)
```
